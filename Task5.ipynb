```python
s#Importing all the libraries that we need.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#importing our dataset.
import pandas as pd
from google.colab import drive
drive.mount('/content/drive')

# Update the path to include your Google Drive location
# Assuming the file is in the 'My Drive' folder of your Google Drive
df = pd.read_csv('/content/drive/My Drive/heart.csv')

# Do Males or Females have higher cholestrol level on average?

# Group the data by sex and calculate the average cholesterol level for each group
average_cholesterol_by_sex = df.groupby('sex')['chol'].mean()

# Print the average cholesterol levels for males and females
print(average_cholesterol_by_sex)

# Create a bar chart to visualize the difference
average_cholesterol_by_sex.plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Average Cholesterol Level by Sex')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.ylabel('Average Cholesterol Level')
plt.xticks(rotation=0)  # Keep x-axis labels horizontal
plt.show()
```

```python
# Calculate the proportion of heart disease for each gender
gender_heart_disease_proportion = df.groupby('sex')['target'].mean()

# Print the proportions
print(gender_heart_disease_proportion)

# Create a bar chart to visualize the proportions
gender_heart_disease_proportion.plot(kind='bar', color=['bisque', 'navy'])
plt.title('Proportion of Heart Disease by Gender')
plt.xlabel('Gender (0 = Female, 1 = Male)')
plt.ylabel('Proportion of Heart Disease')
plt.xticks(rotation=0)
plt.show()
```

```python
# Filter the DataFrame to include only people with heart disease (target = 1)
heart_disease_df = df[df['target'] == 1]

# Calculate the average cholesterol level for people with heart disease
average_cholesterol_heart_disease = heart_disease_df['chol'].mean()

# Print the average cholesterol level
print("Average Cholesterol Level for People with Heart Disease:", average_cholesterol_heart_disease)

# Create a box plot to visualize the cholesterol distribution for people with heart disease
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='chol', data=df, showmeans=True)  # showmeans=True adds mean markers
plt.title('Cholesterol Distribution for People with and without Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Cholesterol Level')
plt.show()
```

```python
# Group data by 'thalach' and calculate the proportion of heart disease for each group
thalach_heart_disease_proportion = df.groupby('thalach')['target'].mean()

# Find the 'thalach' type with the highest proportion of heart disease
most_associated_thalach = thalach_heart_disease_proportion.idxmax()

# Print the most associated 'thalach' type
print("Thalach Type Most Associated with Heart Disease:", most_associated_thalach)

# Highlight the most associated 'thalach' type on the bar chart (optional)
plt.figure(figsize=(10, 6))
plt.bar(thalach_heart_disease_proportion.index, thalach_heart_disease_proportion.values)
plt.bar(most_associated_thalach, thalach_heart_disease_proportion[most_associated_thalach], color='red')  # Highlight in red
plt.title('Proportion of Heart Disease by Thalach Type (Most Associated Highlighted)')
plt.xlabel('Thalach Type (Maximum Heart Rate Achieved)')
plt.ylabel('Proportion of Heart Disease')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
